## Demo: How to work with database


Download database file.


```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

column = 'ConvertedComp'
query = 'SELECT {} FROM master'.format(column)
result = pd.read_sql(query, conn)

result.plot(kind='hist', figsize=(8, 5), legend=False)
plt.title('Distribution Plot of {}'.format(column))
plt.xlabel(column)
```

### Box Plots


Plot a box plot of `Age.`


```python
column = 'Age'
query = 'SELECT {} FROM master'.format(column)
result = pd.read_sql(query, conn)

result.plot(kind='box', figsize=(8, 5), legend=False, vert=False, yticks=[])
plt.title('Box Plot of {}'.format(column))
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
column1 = 'Age'
column2 = 'WorkWeekHrs'
query = 'SELECT {}, {} FROM master'.format(column1, column2)
result = pd.read_sql(query, conn)

result.plot(kind='scatter', x=column1, y=column2, figsize=(18, 8), color='darkblue', legend=False)
plt.title('Scatter Plot of {} vs {}'.format(column1, column2))
plt.xlabel(column1)
plt.ylabel(column2)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import seaborn as sns
column1 = 'Age'
column2 = 'WorkWeekHrs'
column3 = 'CodeRevHrs'

query = 'SELECT {}, {}, {} FROM master'.format(column1, column2, column3)
result = pd.read_sql(query, conn)
result[column3] = result[column3].astype('float')
result = result[result[column2] < 1000]

plt.figure(figsize=(18, 8))
sns.scatterplot(data=result, x=column2, y=column3, size=column1 , sizes=(10,1000))
plt.title('Scatter Plot of {} vs {} by {}'.format(column2, column3, column1))
plt.xlabel(column2)
plt.ylabel(column3)
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
table = 'DatabaseDesireNextYear'
query = '''SELECT DatabaseDesireNextYear, SUM(Respondent) as Sum_Respondents 
            FROM DatabaseDesireNextYear
            GROUP BY 1
            ORDER BY 2 DESC
            LIMIT 5
            '''
result = pd.read_sql(query, conn)
result
```

```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink']
result['Sum_Respondents'].plot(kind='pie', figsize=(15,6), autopct='%1.2f%%', labels=None, shadow=True, colors=colors)
plt.axis('equal')
plt.ylabel('')
plt.legend(labels=result['DatabaseDesireNextYear'])
plt.title('Pie Chart of The Top 5 Databases to Learn Next Year')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
column1 = 'Age'
column2 = 'WorkWeekHrs'
column3 = 'CodeRevHrs'

query = 'SELECT {0}, {1}, {2} FROM master WHERE {0} BETWEEN 30 and 35'.format(column1, column2, column3)
result = pd.read_sql(query, conn)
group = result.groupby('Age').median()

group.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Chart of Medians of WorkWeekHrs and CodeRevHrs with Age Between 30 and 35', y=1.05)
plt.ylim((0,52))
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
column1 = 'Age'
column2 = 'ConvertedComp'

query = 'SELECT {0}, {1} FROM master WHERE {0} BETWEEN 45 and 60'.format(column1, column2)
result = pd.read_sql(query, conn)
group = result.groupby('Age').median()

group.plot(figsize=(10, 6), legend=False)
plt.title('Line Plot of the Median of Converted Compensation by Age')
plt.xlabel(column1)
plt.ylabel(column2)
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
column = 'MainBranch'

query = 'SELECT {}, count(*) as Count FROM master GROUP BY 1'.format(column)
result = pd.read_sql(query, conn).set_index('MainBranch')

result.plot(kind='barh', figsize=(10,6), legend=False)
plt.title('Count of Developers by Branch')
plt.xlabel('Count')
```

Close the database connection.


```python
conn.close()
```
